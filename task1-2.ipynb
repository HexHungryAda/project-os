```python
# All modules/libraries needed
import pandas as pd 
import hashlib as hl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/athlete_events.csv")
df.head()
```

```python
aus_df = df[df["NOC"] == "AUS"]
aus_df.info()
```

```python
annonymous_name = aus_df["Name"].apply(lambda x: hl.sha256(x.encode()).hexdigest())

aus_df.insert(1, "Annonymous Name", annonymous_name)
aus_df.head(1)
```

```python
# Group the sports and count resp. medals on each sport
grouped_sport = aus_df.groupby("Sport", observed=True)[["Medal"]]
medal_count = grouped_sport.count()
medal_count
```

```python
sorted_medals = medal_count.sort_values(["Medal"], ascending=False)
filter_top_10_sports = sorted_medals[:10]
filter_top_10_sports
```

```python
filter_top_10_sports.plot(kind="bar", color="#FF9A3B")

plt.gca().set_facecolor("#FFF4EB")

plt.title("Top 10 total Medals by Sports")
plt.xlabel("Sports")
plt.ylabel("Number of Medals")
plt.legend()
plt.tight_layout()
plt.show()
```

#### Visualisera antal medaljer per OS

```python
aus_df.head()
```

```python
# This part is just to visualize for better understanding the next cell
medals_df = aus_df.groupby(["Games", "Medal"]).size()
medals_df
```

Source:

https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_stacked.html
https://www.geeksforgeeks.org/create-a-stacked-bar-plot-in-matplotlib/

```python
# unstack "Medal" index and convert it into columns
medals_per_os = aus_df.groupby(["Games", "Medal"]).size().unstack(level=1, fill_value=0)
# level 0 = "Games", level 1 = "Medal"
# fill_value=0 fills the rows with 0 if there are 'NaN'
medals_per_os.tail(11)
```

```python
medals_per_os.plot(kind="bar", stacked=True, figsize=(12, 8), color=["brown", "gold", "silver"], edgecolor="black", linewidth=0.7)

plt.title("Number of Medals per Olympic Games")
plt.xlabel("Olympic Games")
plt.ylabel("Number of Medals")
plt.legend(loc="upper left", title="Medal Type", labels=["Bronze", "Gold", "Silver"])
plt.tight_layout()
plt.show()
```

#### Visualisera medaljfördelning mellan länder i 2-4 sporter

```python
sport_1 = df[df["Sport"] == "Tug-Of-War"]
sport_1.head()
```

```python
# What years was this sport in the Olympics?
sport_1["Year"].unique()
```

```python
# What countries have participated?
sport_1["NOC"].unique()
```

```python
# Bar chart of all medals by country.
medals = sport_1.groupby(["NOC", "Medal"]).size().reset_index(name="Count")

medals_final = medals.pivot(index="NOC", columns="Medal", values="Count").fillna(0)

medals_final.plot(kind="bar", color=["brown", "gold", "silver"])

plt.title("Number of Medals by Country")
plt.xlabel("Country")
plt.ylabel("Number of Medals")
plt.legend(title="Medal Type")
plt.tight_layout()
plt.show()
```

```python
# How many people on each team/How many people won gold medal each year?
gold_medals = sport_1[sport_1["Medal"] == "Gold"]
gold_by_year = gold_medals.groupby("Year").size()
gold_by_year.head()
```

```python
# Function to get number of medals by sport
def total_medals_by_country(df, sport):
    what_sport = df[df["Sport"] == sport]
    medals = what_sport.groupby(["NOC", "Medal"]).size().unstack(fill_value=0)
    return medals

total_medals_by_country(df, "Tug-Of-War")
```
